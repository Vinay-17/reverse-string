```python
number=int(input("Enter your number:"))
i=0
even_count=0
odd_count=0
while i<=number:
    if(i%2==0):
        even_count+=1
    else:
        odd_count+=1
    i+=1    
print("Number of EVEN numbers:",even_count)
print("Number of ODD numbers:",odd_count)
output:
Enter your number:54
Number of EVEN numbers: 28
Number of ODD numbers: 27
```
